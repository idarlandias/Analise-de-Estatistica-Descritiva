```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('ACL_sample_1000.csv')

print("Dados carregados com sucesso!")
df.head()
```

```python
# CÓDIGO PARA A QUESTÃO 1

print("--- Análise da Idade (age) ---")

# Calcular a média e a mediana
media_idade = df['age'].mean()
mediana_idade = df['age'].median()

# Exibir os resultados
print(f"Média da idade: {media_idade:.2f}")
print(f"Mediana da idade: {mediana_idade:.2f}")

# Gerar o gráfico para visualização
sns.histplot(df['age'], kde=True, bins=30).set_title('Distribuição de Idade (Age)')
plt.axvline(media_idade, color='red', linestyle='--', label=f'Média')
plt.axvline(mediana_idade, color='green', linestyle='-', label=f'Mediana')
plt.legend()
plt.show()
```

```python
# CÓDIGO PARA A QUESTÃO 2

print("--- Análise do Peso (weight) ---")

# Calcular a média e a mediana
media_peso = df['weight'].mean()
mediana_peso = df['weight'].median()

# Exibir os resultados
print(f"Média do peso: {media_peso:.2f} lbs")
print(f"Mediana do peso: {mediana_peso:.2f} lbs")
print("Como a média é maior que a mediana, isso indica assimetria à direita.")

# Gerar o gráfico para visualização
sns.histplot(df['weight'], kde=True, bins=40).set_title('Distribuição de Peso (Weight)')
plt.axvline(media_peso, color='red', linestyle='--', label=f'Média')
plt.axvline(mediana_peso, color='green', linestyle='-', label=f'Mediana')
plt.legend()
plt.show()
```

```python
# CÓDIGO PARA A QUESTÃO 3

print("--- Análise da Variabilidade Relativa (CV) ---")

# Calcular o Coeficiente de Variação (CV = Desvio Padrão / Média)
cv_weight = df['weight'].std() / df['weight'].mean()
cv_height = df['height'].std() / df['height'].mean()

# Exibir os resultados em porcentagem
print(f"Coeficiente de Variação (Peso): {cv_weight:.2%}")
print(f"Coeficiente de Variação (Altura): {cv_height:.2%}")
print("\nO CV do peso é MAIOR, portanto a afirmativa é falsa.")
```

```python
# CÓDIGO PARA A QUESTÃO 4

print("--- Análise da Distribuição da Altura (height) ---")

# O CV já foi calculado na questão anterior, mas podemos exibi-lo novamente
cv_height = df['height'].std() / df['height'].mean()
print(f"O baixo CV de {cv_height:.2%} confirma a pequena dispersão.")

# Gerar o gráfico para visualização
sns.histplot(df['height'], kde=True).set_title('Distribuição de Altura (Height)')
plt.show()
```

```python
# CÓDIGO PARA A QUESTÃO 5

print("--- Análise da Correlação entre Peso e Altura ---")

# Calcular o coeficiente de correlação de Pearson
correlacao = df['height'].corr(df['weight'])

# Exibir o resultado
print(f"Coeficiente de Correlação: {correlacao:.3f}")
print("Um valor de 0.463 indica uma correlação positiva e moderada.")

# Gerar o gráfico para visualização
sns.scatterplot(data=df, x='height', y='weight').set_title('Relação entre Altura e Peso')
plt.show()
```
